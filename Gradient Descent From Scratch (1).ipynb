### STEPS:

    - Initialize the Coefficient
    - Calculate the Loss
    - Find derivative of the loss function
    - Calculate the loss again
    - Stop when loss until converges(where loss is less than a threshold or doesnt improve anymore)

### TYPES:

    - Batch Gradient descent: The parameters and loss are updated only after all training points have been iterated upon
    - Mini Batch Gradient descent: Does it in batches
    - Stochastic GD - Does it one data point at a time.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(start = -15, stop = 15, num = 30)

y = np.square(x - 5)


plt.figure(figsize=(8,6))
plt.plot(x,y,color='green')
plt.title('Gradient descent')
plt.xlabel('x', color='red')
plt.ylabel('y', color='red')
plt.show()
```

```python
## Gradient a.k.a Derivative of y w.r.t x

gradient_y = lambda x: 2*(x-5)

## Initial value of x close to zero

current_x_value = 0.1

### Precision tells when to stop the algorithm

precision_value = 0.0001

## max number of iterations that can be done

max_iter = 500

### Current iteration 

current_iter = 0

## learning rate

lr = 0.01

## previous step size = current_x_value -  previous_x_value

previous_step_size = 0.5
```

```python
while previous_step_size > precision_value and current_iter < max_iter:
    prev_x_val = current_x_value
    
    current_x_value = current_x_value - lr * gradient_y(prev_x_val)
    
    previous_step_size = abs(current_x_value - prev_x_val)
    
    current_iter = current_iter + 1
    
    print("Iteration count", current_iter, "the current value of x is", current_x_value)

print("The local minima found at", current_x_value)
```
